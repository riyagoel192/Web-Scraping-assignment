```python
soup = BeautifulSoup(r.content) # If this line causes an error, run 'pip install html5lib' or install html5lib
print(soup.prettify())
```

```python
import pandas as pd

df = pd.DataFrame()
l1=[]
l2=[]
l3=[]
l4=[]
l5=[]
print(df)
```

```python
from os import stat_result
for header in soup.find_all('div',attrs={'class':'_13oc-S'}):
  # print(header,"\n")
  for links in header.find_all('a',attrs={'class':'_1fQZEK'}):
    #print("Product URL:" + links.get('href'))
    for smthing in soup.find_all('div', attrs={'class':'col col-7-12'}):
      #print(smthing,"\n\n")
      for heading in smthing.find_all('div', attrs={'class':'_4rR01T'}):
        print("Product Title: " + heading.text)
        l1.append(heading.text)
        for ratings in smthing.find_all('div', attrs={'class':'gUuXy-'}):
          for stars in ratings.find_all('span', attrs={'class':'_1lRcqv'}):
            print("Stars: " + stars.text)
            l2.append(stars.text)
            for reviews in ratings.find_all('span', attrs={'class':'_2_R_DZ'}):
              print("Number of reviews: " + reviews.text)
              l3.append(reviews.text)
              for features in smthing.find_all('div',attrs={'class':'fMghEO'}):
                print("Features: " + features.text)
                l4.append(features.text)
                for price in header.find_all('div',attrs={'class':'col col-5-12 nlI3QM'}):
                  #print(price.text)
                  for amt in price.find_all('div',attrs={'class':'_3tbKJL'}):
                    for inside in amt.find_all('div',attrs={'class':'_25b18c'}):
                      #print(inside)
                      for money in inside.find_all('div',attrs={'class':'_30jeq3 _1_WHN1'}):
                        print("Amount: " + money.text)
                        l5.append(money.text)
                        print("\n\n")

print(l1)
print(l2)
print(l3)
print(l4)
print(l5)
```

```python
df['Title'] = pd.Series(l1)
df['Stars'] = pd.Series(l2)
df['Reviews'] = pd.Series(l3)
df['Features'] = pd.Series(l4)
df['Amount'] = pd.Series(l5)

print(df)


df.to_csv('file1.csv')
```
